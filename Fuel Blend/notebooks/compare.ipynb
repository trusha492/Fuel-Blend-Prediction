```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error

# Load predictions
lgbm_preds = np.load("lgbm_preds.npy").mean(axis=1)
xgb_preds = np.load("xgb_preds.npy").mean(axis=1)
rg_preds = np.load("rg_preds.npy").mean(axis=1)
nn_preds = np.load("nn_test_preds.npy").mean(axis=1)

# Load y_test
y_test = np.load("y_test.npy")

# If multi-target, use only the first target (or adjust as needed)
if y_test.ndim > 1:
    y_test = y_test[:, 0]

# Ensure prediction and y_test lengths match
min_len = min(len(y_test), len(lgbm_preds), len(xgb_preds), len(rg_preds), len(nn_preds))

# Truncate to same length
y_test = y_test[:min_len]
lgbm_preds = lgbm_preds[:min_len]
xgb_preds = xgb_preds[:min_len]
rg_preds = rg_preds[:min_len]
nn_preds = nn_preds[:min_len]

# Evaluate MAE
mae_scores = {
    "LightGBM": mean_absolute_error(y_test, lgbm_preds),
    "XGBoost": mean_absolute_error(y_test, xgb_preds),
    "RandomForest": mean_absolute_error(y_test, rg_preds),
    "NeuralNet": mean_absolute_error(y_test, nn_preds)
}

# Save to CSV
mae_df = pd.DataFrame(list(mae_scores.items()), columns=["Model", "MAE"])
mae_df.to_csv("ensemble.csv", index=False)

print("✅ MAE scores saved to ensemble.csv")
```
